```python
#listes imbriquées contenant les data de chaque fichier
listfreq=[]
listcap=[]
listcond=[]
for i in range(12):
    f2=open(filedf1.loc[:,'AS_data'][i])
    read2=csv.reader(f2)
    listw=[]
    listc=[]
    listg=[]
    for line in read2:
        listw.append(float(line[0]))
        listc.append(float(line[1]))
        listg.append(float(line[2]))
        del listw[139:]
        del listc[139:]
        del listg[139:]
    listfreq.append(listw)
    listcap.append(listc)
    listcond.append(listg)
#verification
print(listfreq[1][135])
#print(type(listfreq))
#print(listfreq)
       
```
